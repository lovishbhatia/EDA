```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set_palette("pastel")
```

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(list(uploaded.keys())[0])
df.head()
```

```python
print("---- SHAPE ----")
print(df.shape)

print("\n---- INFO ----")
print(df.info())

print("\n---- NULL VALUES ----")
print(df.isnull().sum())
```

```python
df.describe(include='all')
```

UNIVARIATE ANALYSIS

```python
num_cols = df.select_dtypes(include=['float64','int64']).columns

plt.figure(figsize=(15, 12))

for i, col in enumerate(num_cols, 1):
    plt.subplot(4, 4, i)
    sns.histplot(df[col], kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(y=df[col], order=df[col].value_counts().head(20).index)
    plt.title(f"Top values in {col}")
    plt.show()
```

BIVARIATE ANALYSIS

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['danceability'], y=df['popularity'])
plt.title("Popularity vs Danceability")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['energy'], y=df['popularity'])
plt.title("Popularity vs Energy")
plt.show()
```

```python
plt.figure(figsize=(12,6))
top_genres = df['genre'].value_counts().head(15).index
sns.boxplot(data=df[df['genre'].isin(top_genres)],
            x='genre', y='popularity')
plt.xticks(rotation=45)
plt.title("Popularity by Genre")
plt.show()
```

CORRELATION ANALYSIS

```python
plt.figure(figsize=(12,10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
corr_pop = corr['popularity'].sort_values(ascending=False)
corr_pop
```

COUNTRY & YEAR ANALYSIS

```python
plt.figure(figsize=(10,5))
top_countries = df['country'].value_counts().head(10).index
sns.barplot(x='country', y='popularity', data=df[df['country'].isin(top_countries)])
plt.xticks(rotation=45)
plt.title("Popularity by Country")
plt.show()
```

```python
df['release_year'] = pd.to_datetime(df['release_date']).dt.year

plt.figure(figsize=(10,5))
sns.countplot(x='release_year', data=df)
plt.xticks(rotation=45)
plt.title("Songs Released per Year")
plt.show()
```

STREAM COUNT ANALYSIS

```python
plt.figure(figsize=(7,4))
sns.histplot(df['stream_count'], bins=50, kde=True)
plt.title("Stream Count Distribution")
plt.show()
```

```python
sns.scatterplot(x='stream_count', y='popularity', data=df)
plt.title("Popularity vs Stream Count")
plt.show()
```

 TOP SONGS / ARTISTS

```python
plt.figure(figsize=(10,5))
df['artist_name'].value_counts().head(20).plot(kind='bar')
plt.title("Top 20 Artists by Number of Songs")
plt.show()
```
